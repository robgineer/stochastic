# Aufgabe 2.3

Sei $(A_1, A_2, A_3,...)$ eine Familie von Ereignissen auf einem Wahrscheinlichkeitsraum $(\Omega,A, P)$. 
Im  Folgenden  bezeichne

$A^c= \Omega \ A$ das zum Ereignis $A$ komplementäre Ereignis.


Zeigen Sie die folgenden Aussagen.

(a) Für alle $n \in \mathbb{N} $ gilt: 
\begin{equation}
(\cup_{i=1}^n A_i)^c = \cap_{i=1}^n A_i^c
\end{equation}
und
\begin{equation}
(\cap_{i=1}^n A_i)^c = \cup_{i=1}^n A_i^c
\end{equation}

(b) Für alle natürlichen Zahlen $n \geq 2 $ gilt:

\begin{equation}
P(\cup_{i=1}^n A_i ) \geq \sum_{i=1}^n P(A_i) −  \sum_{1 \leq i \leq j \leq n} P(A_i \cap A_j)
\end{equation}

(c) Für alle natürlichen Zahlen $n \geq 2 $ gilt:

\begin{equation}
P(\cup_{i=1}^n A_i ) \leq \sum_{i=1}^n P(A_i) −  \sum_{1 \leq i \leq j \leq n} P(A_i \cap A_j) + \sum_{1 \leq i \leq j \leq k \leq n} P(A_i \cap A_j \cap A_k)
\end{equation}

Hier wird in den Doppelsummen ̈uber alle Indexpaare  $(i, j)$  mit $i, j \in \{1,..., n\}$ und $i < j$ summiert und in der Dreifachsumme ̈uber alle Indextripel  $(i, j, k)$  mit $i, j, k \in \{1,..., n\}$ und $i < j < k$.

## a)
Nach de Morgan gilt:

\begin{equation}
\neg(A \vee B) = \neg A \wedge \neg B
\end{equation}
und 
\begin{equation}
\neg(A \wedge B) = \neg A \vee \neg B
\end{equation}

Damit ist 
\begin{equation}
(\cup_{i=1}^n A_i)^c = \{(A_1 \cup A_2 \cup A_3 \cup ...  \cup A_n)^c\} \Leftrightarrow \{A_1^c \cap A_2^c \cap A_3^c \cap ...  \cap A_n^c\} = \cap_{i=1}^n A_i^c
\end{equation}


\begin{equation}
(\cap_{i=1}^n A_i)^c = \{(A_1 \cap A_2 \cap A_3 \cap ...  \cap A_n)^c\} \Leftrightarrow \{A_1^c \cup A_2^c \cup A_3^c \cup ...  \cup A_n^c\} = \cup_{i=1}^n A_i^c
\end{equation}


## (b) 

\begin{equation}
P(\cup_{i=1}^n A_i ) \geq \sum_{i=1}^n P(A_i) −  \sum_{1 \leq i \le j \leq n} P(A_i \cap A_j)
\end{equation}



\begin{equation}
P(\cup_{i=1}^n A_i ) = P(A_1) \cup P(A_2)  \cup ... \cup P(A_n)
\end{equation}

\begin{equation}
\sum_{i=1}^n P(A_i) = P(A_1) + P(A_2) + ... + P(A_n)
\end{equation}


Annahme: $A_i \ne \{0\}$. Falls $A_i$ paarweise disjunkt sind, so ist nach Satz 5.1.9:

\begin{equation}
P(\cup_{i=1}^n A_i) = \sum_{i=1}^n P(A_i)
\end{equation}

und damit

\begin{equation}
\forall A \in \mathrm{A}: \sum_{1 \leq i < j \leq n} P(A_i \cap A_j) = 0 
\end{equation}


Damit ist die Aussage 

\begin{equation}
P(\cup_{i=1}^n A_i ) \geq \sum_{i=1}^n P(A_i) −  \sum_{1 \leq i < j \leq n} P(A_i \cap A_j) 
\end{equation}

richtig.


Falls $A_i$ nicht paarweise disjunkt sind, so ist 

\begin{equation}
\forall A \in \mathrm{A}: \sum_{1 \leq i < j \leq n} P(A_i \cap A_j) > 0
\end{equation}

Nach Satz 5.1.9 gilt:

\begin{equation}
P(\cup_{i=1}^n A_i ) \leq \sum_{i=1}^n P(A_i)
\end{equation}

und damit

\begin{equation}
P(\cup_{i=1}^n A_i ) > \sum_{i=1}^n P(A_i) −  \sum_{1 \leq i < j \leq n} P(A_i \cap A_j) 
\end{equation}
